```python
X = df.drop(columns=['date', 'lights', 'Appliances'], axis = 1)
Y= df['Appliances']
```

```python
#train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#feature scaling
scaler = MinMaxScaler()
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test) 
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(rescaledX_train, y_train)
```

```python
pred_linear = linreg.predict(rescaledX_test)
pred_linear
```

```python
mae = round(mean_absolute_error(y_test, pred_linear), 2)
mse = round(mean_squared_error(y_test, pred_linear), 2)
print(f'The mean absolute error for the linear regression model is {mae}')
print(f'The mean squared error for the linear regression model is {mse}')
```

```python
r2_score = round(linreg.score(rescaledX_test, y_test), 2)
print(f'The r2_score for the linear regression model is {r2_score}')
```

```python
weights_linear = list(linreg.coef_)
data = list(zip(X_train.columns, weights_linear))
weights_linear_df = pd.DataFrame(data, columns=['Features', 'Weights']).sort_values(by='Weights', ascending=False)
weights_linear_df
```

```python
l1 = Lasso(alpha=0.001)
```

```python
l1.fit(rescaledX_train, y_train)
```

```python
pred_l1 = l1.predict(rescaledX_test)
```

```python
l1_score = round(l1.score(rescaledX_test, y_test),2)
print(f'The r2 score for lasso is {l1_score}')
```
